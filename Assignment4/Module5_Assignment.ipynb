```python
import pandas as pd
import numpy as np
import seaborn as sns 
#import matplotlib as mpl
from matplotlib import pyplot as plt

teamsdf = pd.read_csv('https://raw.githubusercontent.com/frankData612/data_612/master/baseballdatabank-master/core/Teams.csv')

teamsdf.head()
```

```python
teamsdf.info()
```

```python
teamsdf.describe()
```

```python
fig = sns.pairplot(teamsdf.iloc[:,14:20])
```

```python
box, ax = plt.subplots()
ax = sns.boxplot(x='Rank', y='W', data=teamsdf)
ax.set_title('Boxplot of Team Rank by Wins')
ax.set_xlabel('Team Rank')
ax.set_ylabel('Numer of Wins')
plt.show()
```

```python
teamsdf['franchID'].unique()
```

```python
dataset_1 = teamsdf[teamsdf['franchID']=='DET']
dataset_2 = teamsdf[teamsdf['franchID']=='CLE']
dataset_3 = teamsdf[teamsdf['franchID']=='TOR']
dataset_4 = teamsdf[(teamsdf['divID']=='C')]
dataset_5 = teamsdf[teamsdf['lgID']=='AL']
```

```python
hist, ax = plt.subplots()
ax = sns.histplot(dataset_1['FP'])
ax.set_title("Detriot Tigers Fielding Percentage Histogram")
plt.show()
```

```python
bar, ax=plt.subplots()
ax = sns.barplot(x='LgWin', y='CG', data=dataset_5)
ax.set_title('Bar chart of League Wins for Completed Games in AL league')
ax.set_xlabel('League Wins')
ax.set_ylabel('Completed Games')
plt.show()
```
