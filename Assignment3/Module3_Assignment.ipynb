```
import pandas as pd
from datetime import datetime as d
from datetime import timedelta as t
import numpy as np

df = pd.read_csv('https://raw.githubusercontent.com/frankData612/data_612/master/stock_data/stocks_yahoo.csv')

df['date_dt'] = pd.to_datetime(df['date'], format='%b %d %Y')

dt = df['date_dt']

maxdate = dt.max()

day_age = (maxdate-df['date_dt'])

#year = day_age/365
#month = year*12
#nm = pd.to_datetime(month, format='%M')

#type(maxdate)
#type(df['date'])

df['no of Months'] = day_age/np.timedelta64(1,'M')

df.head(20)
```
